```python
#ADDING EVENTS
typetodbname = {"50/50":"fifty_fifties", 
                "Bad Behaviour":"bad_behaviours", 
                "Ball Receipt*":"ball_receipts", 
                "Ball Recovery":"ball_recoveries",  
                "Block":"blocks", 
                "Carry":"carries", 
                "Clearance":"clearances",
                "Dispossessed":"dispossesseds",   
                "Dribble":"dribbles", 
                "Dribbled Past":"dribbles_past",
                "Duel":"duels",
                "Error":"errors", 
                "Foul Committed":"fouls_committed",
                "Foul Won":"fouls_won",  
                "Goal Keeper":"goalkeepers",
                "Half End":"half_ends",
                "Half Start":"half_starts", 
                "Injury Stoppage":"injury_stoppages",
                "Interception":"interceptions", 
                "Miscontrol":"miscontrols", 
                "Offside":"offsides",
                "Own Goal Against":"own_goals_against",
                "Own Goal For":"own_goals_for",
                "Pass":'passes', 
                "Player Off":"players_off", 
                "Player On":"players_on", 
                "Pressure":"pressures", 
                "Referee Ball-Drop":"ref_ball_drops",
                "Shield":"shields",
                "Shot":"shots",
                #"Starting XI":"startingxis",
                "Substitution":"substitutions",
                "Tactical Shift":"tactical_shifts",
                "Event":"events"
                }
#stores events post-processing
allevents = {}
for typename in typetodbname:
    allevents[typename] = []

def boolLambdas(name):
    return lambda x: x[name] if name in x else False
def otherLambdas(name):
    return lambda x: x[name] if name in x else None
def nameNesteds(name):
    return lambda x: x[name]['name'] if name in x else None
#used to process events into insertable dict form
shot_cols = {"key_pass_id":otherLambdas('key_pass_id'),
             "endloc_x":lambda x: x['end_location'][0] if 'end_location' in x else None,
            "endloc_y":lambda x: x['end_location'][1] if 'end_location' in x else None,
            "endloc_z":lambda x: x['end_location'][2] if 'end_location' in x and len(x['end_location'])>2 else None,
            "aerial_won":boolLambdas('aerial_won'),
            "follows_dribble":boolLambdas('follows_dribble'),
            "first_time":boolLambdas("first_time"),
            "open_goal":boolLambdas("open_goal"),
            "xgoals":otherLambdas("statsbomb_xg"),
            "deflected":boolLambdas("deflected"),
            "outcome":nameNesteds('outcome'),
            "technique":nameNesteds('technique'),
            "body_part":nameNesteds('body_part'),
            "shot_type":nameNesteds('type')}
pass_cols = {"recipient":lambda x: x['recipient']['id'] if ('recipient' in x) else None,
                "pass_length":otherLambdas('length'),
                "angle":otherLambdas('angle'),
                "height":nameNesteds('height'),
                "endloc_x":lambda x: x['end_location'][0] if 'end_location' in x else None,
                "endloc_y":lambda x: x['end_location'][1] if 'end_location' in x else None,
                "assisted_shot_id":otherLambdas('assisted_shot_id'),
                "backheel":boolLambdas('backheel'),
                "deflected":boolLambdas('deflected'),
                "miscommunication":boolLambdas('miscommunication'),
                "was_cross":boolLambdas('cross'),
                "cutback":boolLambdas('cut_back'),
                "switch":boolLambdas('switch'),
                "shot_assist":boolLambdas('shot_assist'),
                "goal_assist":boolLambdas('goal_assist'),
                "pass_type":nameNesteds('type'),
                "outcome":nameNesteds('outcome'),
                "technique":nameNesteds('technique')}
dribbles_cols = {"overrun":boolLambdas('overrun'),
                 "nutmeg":boolLambdas('nutmeg'),
                 "notouch":boolLambdas('no_touch'),
                 "complete":lambda x: True if 'outcome' in x and x['outcome']['name'] == "Complete" else False}
#dribbled_past_cols = {"counterpress":"idkyet"}
fifty_cols = {"outcome":nameNesteds('outcome'),
              #"counterpress":"idk yet"
              }
badbehaviour_cols = {"card_type":nameNesteds('card')}
ballreceipt_cols = {"complete":lambda x: False if 'outcome' in x and x['outcome']['name'] == "Incomplete" else True}
ballrecovery_cols = {"offensive":boolLambdas('offensive'),
                     "recovery_failure":boolLambdas('recovery_failure')}
block_cols = {"deflection":boolLambdas('deflection'),
              "offensive":boolLambdas('offensive'),
              "save_block":boolLambdas('save_block'),
              #"counterpress":'idkyet'
              }
carry_cols = {"endloc_x":lambda x: x['end_location'][0] if 'end_location' in x else None,
                "endloc_y":lambda x: x['end_location'][1] if 'end_location' in x else None,}
clearance_cols = {"aerial_won":boolLambdas('aerial_won'),
                  "body_part":nameNesteds('body_part')}
duel_cols = {#"counterpress":"idkyet",
             "duel_type":nameNesteds('type'),
             "outcome":nameNesteds('outcome')}
foulscommitted_cols = {#"counterpress":"idkyet",
                       "offensive":boolLambdas('offensive'),
                       "foul_type":nameNesteds('type'),
                       "advantage":boolLambdas('advantage'),
                       "penalty":boolLambdas('penalty'),
                       "card":nameNesteds('card')}
foulswon_cols = {"defensive":boolLambdas('defensive'),
                 "advantage":boolLambdas('advantage'),
                 "penalty":boolLambdas('penalty')}
goalkeeper_cols = {"position":nameNesteds('position'),
                   "technique":nameNesteds('technique'),
                   "body_part":nameNesteds('body_part'),
                   "gk_type":nameNesteds('type'),
                   "outcome":nameNesteds('outcome')}
halfends_cols = {"early_vid_end":boolLambdas('early_video_end'),
                 "suspended":boolLambdas('suspended')}
halfstarts_cols = {"late_vid_start":boolLambdas("late_video_start")}
injury_stoppages_cols = {"in_chain":boolLambdas("in_chain")}
interception_cols = {"outcome":nameNesteds("outcome")}
miscontrols_cols = {"aerial_won":boolLambdas('aerial_won')}
playeroff_cols = {"permanent":boolLambdas('permanent')}
#pressure_cols = {"counterpress":"idkyet"}
substitution_cols = {"replacement_player":nameNesteds('replacement'),
                     "outcome":nameNesteds('outcome')}
tactical_cols = {"formation":otherLambdas('formation')}
eventcols = {"Shot":shot_cols,
             "Pass":pass_cols,
             "Dribble":dribbles_cols,
             #"Dribbled Past":dribbled_past_cols,
             "50/50":fifty_cols,
             "Bad Behaviour":badbehaviour_cols,
             "Ball Receipt*":ballreceipt_cols,
             "Ball Recovery":ballrecovery_cols,
             "Block":block_cols,
             "Carry":carry_cols,
             "Clearance":clearance_cols,
             "Duel":duel_cols,
             "Foul Committed":foulscommitted_cols,
             "Foul Won":foulswon_cols,
             "Goal Keeper":goalkeeper_cols,
             "Half End":halfends_cols,
             "Half Start":halfstarts_cols,
             "Injury Stoppage":injury_stoppages_cols,
             "Interception":interception_cols,
             "Miscontrol":miscontrols_cols,
             "Player Off":playeroff_cols,
             "Substitution":substitution_cols,
             "Tactical Shift":tactical_cols}

def processEvent(e, cols):
    newE = {}
    for col in cols:
        newE[col] = cols[col](e)
    return newE

def makeQueryString(events, table):
    s = "INSERT INTO " + table + "("
    for col in events[0]:
        s = s+col+","
    s = s[:-1]+") VALUES("
    for e in events:
        for col in e:
            if type(e[col]) is str: s = s+"'"+e[col]+"', "
            elif e[col] is None: s = s+"NULL, "
            else: s = s+str(e[col])+", "
        s = s[:-2]+"), ("
    s = s[:-3]
    return s


#finish columns dictionary for each type
#make dictionary of column dictionaries with e_type as key
#loop through all events
#get corresponding col dictionary based on type, run processevent, add to all events
#once done looping, run executemany for each type in allevents

import json, os, psycopg, math

matches_path = 'open-data/data/matches/'
events_path = 'open-data/data/events/'
directory = os.fsencode(matches_path)

#la liga is comp_id 11, epl is 2
la_liga = 11
epl = 2

epl_seasons = [44]
la_liga_seasons = [4, 42, 90]
eplx = '44.json'

with psycopg.connect("dbname=project_database user=postgres password=postgres") as conn:
    with conn.cursor() as cur:
        #Going over the competition directories
        for comp_dir in os.listdir(directory):
            comp_id = os.fsdecode(comp_dir)
            if comp_id == str(la_liga) or comp_id == str(epl):
                d = os.fsencode(matches_path+comp_id)
                #going over the seasons for each competition
                for season_file in os.listdir(d):
                    season_id = os.fsdecode(season_file)
                    if (comp_id == str(la_liga) and season_id in [str(item)+'.json' for item in la_liga_seasons]) or (comp_id == str(epl) and season_id == eplx):
                        match_file_contents = open(matches_path+comp_id+'/'+season_id, "r", encoding="utf-8")
                        matches_json = json.loads(match_file_contents.read())
                        for game in matches_json:
                            match_id = game['match_id']
                            eventfilecontents = open(events_path+str(match_id)+".json", "r", encoding='utf-8')
                            event_json = json.loads(eventfilecontents.read())
                            for event in event_json:
                                parsedE = {}
                                parsedE['event_id'] = event['id']
                                parsedE['match_id'] = match_id
                                parsedE['event_index'] = event['index']
                                parsedE['match_period'] = event['period']
                                parsedE['time_start'] = event['timestamp']
                                parsedE['minut'] = event['minute']
                                parsedE['sec'] = event['second']
                                parsedE['possession'] = event['possession']
                                parsedE['poss_team'] = event['possession_team']['id']
                                parsedE['play_pattern'] = event['play_pattern']['name']
                                parsedE['team_id'] = event['team']['id']
                                parsedE['team_name'] = event['team']['name']
                                if 'player' in event:
                                        parsedE['player_id'] = event['player']['id']
                                        parsedE['player_name'] = event['player']['name'].replace("'", "''")
                                else:
                                    parsedE['player_id'] = None
                                    parsedE['player_name'] = None
                                if 'position' in event: 
                                    parsedE['player_pos'] = event['position']['name']
                                else: 
                                    parsedE['player_pos'] = None
                                if 'under_pressure' in event: 
                                    parsedE['under_pressure'] = event['under_pressure']
                                else: 
                                    parsedE['under_pressure'] = False
                                if 'out' in event: 
                                    parsedE['goes_out'] = event['out']
                                else: 
                                    parsedE['goes_out'] = False
                                if 'duration' in event: 
                                    parsedE['duration'] = event['duration']
                                else: 
                                    parsedE['duration'] = None
                                if 'tactics' in event: 
                                    parsedE['tactics'] = event['tactics']['formation']
                                else: 
                                    parsedE['tactics'] = None
                                if 'location' in event:
                                    parsedE['loc_x'] = event['location'][0]
                                    parsedE['loc_y'] = event['location'][1]
                                else:
                                    parsedE['loc_x'] = parsedE['loc_y'] = None
                                if 'counterpress' in event:
                                    parsedE['counterpress'] = event['counterpress']
                                else:
                                    parsedE['counterpress'] = False
                                etype = event['type']['name']
                                if etype in eventcols: 
                                    if etype == "Goal Keeper": eObjName = 'goalkeeper'
                                    elif etype == "Ball Receipt*": eObjName = 'ball_receipt'
                                    else: eObjName = etype.lower().replace(' ', '_')
                                    if eObjName in event: 
                                        eventObj = event[eObjName]
                                        p = processEvent(eventObj, eventcols[event['type']['name']])
                                        parsedE.update(p)
                                    else:
                                        eventObj = {}
                                        p = processEvent(eventObj, eventcols[event['type']['name']])
                                        parsedE.update(p)
                                if etype in allevents: allevents[etype].append(parsedE)   

        for table in allevents:
            for i in range(math.ceil(len(allevents[table])/250)):
                es = allevents[table][(i*250):(i+1)*250]
                try:
                    print(table+""+str(i*250))
                    q = makeQueryString(es, typetodbname[table])
                    
                    cur.execute(q)
                except psycopg.IntegrityError:
                    conn.rollback()
                else:
                    conn.commit()
                                
```
